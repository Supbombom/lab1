#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
# Your code goes here
def count_vowels(text):
    vowels = {'a', 'e', 'i', 'o', 'u'}
    
    # текст к нижнему регистру
    text_lower = text.lower()

    count = 0
    for char in text_lower:
        if char in vowels:
            count += 1
            
    return count

class TestCountVowels(unittest.TestCase):
    
    def test_only_vowels(self):
        """Тест строки только из гласных"""
        self.assertEqual(count_vowels("aeiou"), 5)
        self.assertEqual(count_vowels("aei"), 3)
    
    def test_only_consonants(self):
        """Тест строки только из согласных"""
        self.assertEqual(count_vowels("bcdfg"), 0)
        self.assertEqual(count_vowels("xyz"), 0)
    
    def test_uppercase_letters(self):
        """Тест с заглавными буквами"""
        self.assertEqual(count_vowels("AEIOU"), 5)
        self.assertEqual(count_vowels("Hello WORLD"), 3)

test_suite = unittest.TestLoader().loadTestsFromTestCase(TestCountVowels)
unittest.TextTestRunner(verbosity=2).run(test_suite)
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
# Your code goes here
def has_unique_chars(text):
    return len(text) == len(set(text))

#Тысты
class TestHasUniqueChars(unittest.TestCase):
    
    def test_unique_characters(self):
        """Тест строк с уникальными символами"""
        self.assertTrue(has_unique_chars("abc"))
        self.assertTrue(has_unique_chars("123"))
        self.assertTrue(has_unique_chars("aBcD"))
    
    def test_duplicate_characters(self):
        """Тест строк с повторяющимися символами"""
        self.assertFalse(has_unique_chars("hello"))
        self.assertFalse(has_unique_chars("112233"))
        self.assertFalse(has_unique_chars("  "))
     

test_suite = unittest.TestLoader().loadTestsFromTestCase(TestHasUniqueChars)
unittest.TextTestRunner(verbosity=2).run(test_suite)
```

бит равных 1 в этом числе. Написать тесты для кода

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    # Your code goes here
    if n < 10:
        return 0
    
    count = 0
    current_number = n
    
    while current_number >= 10:
        digits = [int(digit) for digit in str(current_number)]
        
        product = 1
        for digit in digits:
            product *= digit
        
        current_number = product
        count += 1
    
    return count
#Тесты
class TestMagic(unittest.TestCase):
 def test_1(self):
    self.assertEqual(magic(39), 3)
    self.assertEqual(magic(4), 0)
    self.assertEqual(magic(999), 4)
    self.assertEqual(magic(25), 2)  
    self.assertEqual(magic(10), 1)    
    self.assertEqual(magic(123), 1)

test_suite = unittest.TestLoader().loadTestsFromTestCase(TestMagic)
unittest.TextTestRunner(verbosity=2).run(test_suite)
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    # Your code goes here
    n = len(pred)
    total = 0
    
    # cуммируем квадраты разностей
    for i in range(n):
        difference = pred[i] - true[i]
        total += difference ** 2
    
    # среднее значение квадратов разностей
    return total / n   

#Тесты
class TestMSE(unittest.TestCase):
    
    def test_identical_vectors(self):
        """Тест одинаковых векторов (MSE = 0)"""
        pred = [1, 2, 3, 4, 5]
        true = [1, 2, 3, 4, 5]
        self.assertEqual(mse(pred, true), 0.0)

    def test_float_values(self):
        """Тест с дробными числами"""
        pred = [1.5, 2.5, 3.5]
        true = [1.0, 2.0, 3.0]
        self.assertAlmostEqual(mse(pred, true), 0.25)

    def test_constant_difference(self):
        """Тест постоянной разницы"""
        pred = [2, 3, 4, 5, 6]  
        true = [1, 2, 3, 4, 5]
        self.assertEqual(mse(pred, true), 1.0) 

test_suite = unittest.TestLoader().loadTestsFromTestCase(TestMSE)
unittest.TextTestRunner(verbosity=2).run(test_suite)   
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
# Your code goes here
def prime_factors(n):
    if n <= 1:
        return f"({n})"
    
    factors = []
    divisor = 2
    
    # Разлагаем число на простые множители
    while divisor * divisor <= n:
        count = 0
        while n % divisor == 0:
            count += 1
            n //= divisor
        if count > 0:
            factors.append((divisor, count))
        divisor += 1
    
    # Если остался простой делитель больше 1
    if n > 1:
        factors.append((n, 1))
    
    # Формируем строку результата
    result = []
    for factor, power in factors:
        if power == 1:
            result.append(f"({factor})")
        else:
            result.append(f"({factor}**{power})")

    return "".join(result)

# Tests
class TestPrimeFactors(unittest.TestCase):
    
    def test_prime_numbers(self):
        """Тест простых чисел"""
        self.assertEqual(prime_factors(2), "(2)")
        self.assertEqual(prime_factors(7), "(7)")
        self.assertEqual(prime_factors(13), "(13)")
    
    def test_powers_of_primes(self):
        """Тест степеней простых чисел"""
        self.assertEqual(prime_factors(4), "(2**2)")
        self.assertEqual(prime_factors(8), "(2**3)")
        self.assertEqual(prime_factors(9), "(3**2)")
        self.assertEqual(prime_factors(25), "(5**2)")

    def test_numbers_with_many_factors(self):
        """Тест остальных чисел"""
        self.assertEqual(prime_factors(2310), "(2)(3)(5)(7)(11)")
        self.assertEqual(prime_factors(1024), "(2**10)")

test_suite = unittest.TestLoader().loadTestsFromTestCase(TestPrimeFactors)
unittest.TextTestRunner(verbosity=2).run(test_suite)   
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    # Your code goes here
    if number <= 0:
        return "It is impossible"
    
    k = 0
    total = 0
    
    while total < number:
        k += 1
        total += k * k
        
        if total == number:
            return k
    
    return "It is impossible"    

#Tests
class TestPyramid(unittest.TestCase):
    
    def test_small_pyramids(self):
        """Тест пирамид"""
        self.assertEqual(pyramid(1), 1)
        self.assertEqual(pyramid(5), 2)
        self.assertEqual(pyramid(14), 3)
        self.assertEqual(pyramid(30), 4)
    
    def test_impossible_numbers(self):
        """Тест чисел, из которых нельзя построить пирамиду"""
        self.assertEqual(pyramid(2), "It is impossible")
        self.assertEqual(pyramid(3), "It is impossible")
        self.assertEqual(pyramid(6), "It is impossible")
        self.assertEqual(pyramid(7), "It is impossible")
        self.assertEqual(pyramid(15), "It is impossible")
        self.assertEqual(pyramid(0), "It is impossible")
        self.assertEqual(pyramid(-5), "It is impossible")

test_suite = unittest.TestLoader().loadTestsFromTestCase(TestPyramid)
unittest.TextTestRunner(verbosity=2).run(test_suite)   
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
# Your code goes here
def is_balanced(number):

    digits = str(number)
    n = len(digits)
    if n == 1:
        return True

    if n % 2 == 0:
        mid_start = n // 2 - 1
        mid_end = n // 2 + 1
    else:
        mid_start = n // 2
        mid_end = n // 2 + 1
    
    # Вычисляем сумму цифр до средних и после средних
    sum_before = sum(int(digit) for digit in digits[:mid_start])
    sum_after = sum(int(digit) for digit in digits[mid_end:])
    
    return sum_before == sum_after

class TestIsBalanced(unittest.TestCase):
    def test_even_length_balanced(self):
        """Тест сбалансированных чисел с четной длиной"""
        self.assertTrue(is_balanced(1221)) 
        self.assertTrue(is_balanced(123321))  
    
    def test_even_length_unbalanced(self):
        """Тест несбалансированных чисел с четной длиной"""
        self.assertFalse(is_balanced(1234))
        self.assertFalse(is_balanced(1123))
        self.assertFalse(is_balanced(9991))
    
    def test_odd_length_balanced(self):
        """Тест сбалансированных чисел с нечетной длиной"""
        self.assertTrue(is_balanced(121))
        self.assertTrue(is_balanced(12321))
        self.assertTrue(is_balanced(13531))
        self.assertTrue(is_balanced(9876789))
    
    def test_odd_length_unbalanced(self):
        """Тест несбалансированных чисел с нечетной длиной"""
        self.assertFalse(is_balanced(123))
        self.assertFalse(is_balanced(12345))
        self.assertFalse(is_balanced(11121))

test_suite = unittest.TestLoader().loadTestsFromTestCase(TestIsBalanced)
unittest.TextTestRunner(verbosity=2).run(test_suite)   
```
